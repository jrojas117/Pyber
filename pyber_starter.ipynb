```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_city_ride = pd.merge(ride_data_df,city_data_df, on="city", how="left")

# Display the data table for preview
merge_city_ride.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Separate from the merge data frame the different types of city and create a new data frame with each of them
urban_city_type = merge_city_ride[merge_city_ride["type"] == "Urban"]
rural_city_type = merge_city_ride[merge_city_ride["type"] == "Rural"]
suburban_city_type = merge_city_ride[merge_city_ride["type"] == "Suburban"]
#group the new data frames by city 
urban_city_type = urban_city_type.groupby("city")
rural_city_type = rural_city_type.groupby("city")
suburban_city_type = suburban_city_type.groupby("city")

# Obtain the x and y coordinates for each of the three city types
x_urban = urban_city_type["ride_id"].count()
y_urban = urban_city_type["fare"].mean()
z_urban = urban_city_type["driver_count"].max() #this will give us the size of the bubble

x_rural = rural_city_type["ride_id"].count()
y_rural = rural_city_type["fare"].mean()
z_rural = rural_city_type["driver_count"].max() #this will give us the size of the bubble

x_suburban = suburban_city_type["ride_id"].count()
y_suburban = suburban_city_type["fare"].mean()
z_suburban = suburban_city_type["driver_count"].max() #this will give us the size of the bubble

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, label = "Urban", s=z_urban * 10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o",zorder=3)
plt.scatter(x_rural, y_rural, label = "Rural", s=z_rural * 10, color=["gold"], edgecolor="black", alpha = 0.75, marker="o",zorder=3)
plt.scatter(x_suburban, y_suburban, label = "Suburban", s=z_suburban * 10, color=["lightskyblue"], edgecolor="black", alpha = 0.75, marker="o",zorder=3)
plt.grid(zorder=0)

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend
legend = plt.legend(fontsize = 9, title= "City Types", loc="best")
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,40,"Note: \nCircle size correlates with \ndriver count per city.", fontsize = 10)

# Save Figure
plt.savefig("Pyber.png", bbox_inches="tight")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
merge_data_bytype = merge_city_ride.groupby(['type'])
# Calculate Type Percents
Total_Fares = merge_data_bytype['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "coral"]
plt.pie(Total_Fares, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("TotalFaresbyCityType.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
Total_Rides = merge_data_bytype['ride_id'].count()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "coral"]
plt.pie(Total_Rides, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("TotalFaresbyCityType.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
Total_Drivers = city_data_df.groupby(['type']).sum()
# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "coral"]
plt.pie(Total_Drivers, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("TotalFaresbyCityType.png")

# Show Figure
plt.show()
```
